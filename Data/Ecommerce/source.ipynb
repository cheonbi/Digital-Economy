```python
# 데이터 로딩
user_ecommerce_purchase = pd.read_csv('./ecommerce_user.csv')
# 데이터 2개로 분리
A_col = ['IP Address', 'Address', 'Email', 'Job', 'Lot']
B_col = ['IP Address', 'Purchase Price', 'AM or PM', 'Browser Info', 'Language', 'Company', 'Credit Card', 'CC Exp Date', 'CC Security Code', 'CC Provider']
user_info = user_ecommerce_purchase[A_col].copy()
user_purchase = user_ecommerce_purchase[B_col].copy()
# 분리결과 출력
display(user_info, user_purchase)
# 데이터 조작 위한 분리
user_info_new = user_info.sample(int(len(user_info)*0.85)).reset_index().iloc[:,1:]
df_sub_sample = user_purchase.sample(5148, random_state=123)
index_new = [i for i in user_purchase.index if i not in df_sub_sample.index]
df_sub = user_purchase.loc[index_new]
# fake ip 생성 및 반영
from faker import Faker
fake_ip = [Faker().ipv4() for i in range(df_sub_sample.shape[0])]
df_sub_sample['IP Address'] = fake_ip
# 최종 데이터 생성
user_purchase_new = pd.concat([df_sub, df_sub_sample], axis=0).sort_index()
# 컬럼명 변경
user_purchase_new = user_purchase_new[['Purchase Price', 'AM or PM', 'Browser Info', 'Language',
                                      'Company', 'Credit Card', 'CC Exp Date', 'CC Security Code',
                                      'CC Provider', 'IP Address']].copy()
if 'IP Address' in user_purchase_new.columns:
    user_purchase_new.rename(columns = {'IP Address' : 'Address'}, inplace = True)
display(user_info_new, user_purchase_new)
# 저장
user_info_new.to_csv('ecommerce_userinfo.csv', index=False)
user_purchase_new.to_csv('ecommerce_userpurchase.csv', index=False)
```
